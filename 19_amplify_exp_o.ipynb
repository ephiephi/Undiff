```python
#### Display tools
def plot_this(s,title=''):
    """

    """
    import pylab
    s = s.squeeze()
    if s.ndim ==1:
        pylab.plot(s)
    else:
        pylab.imshow(s,aspect='auto')
        pylab.title(title)
    pylab.show()
```

```python
origin = "/data/ephraim/datasets/known_noise/undiff_exps/exp_o_net3_6_copy/"
origins = [origin + "noisy_wav/book_hfchannel_snr5_fileid_18.wav",
         origin + "noisy_wav/book_buccaneer1_snr5_fileid_6.wav",
         origin + "noises/noise_fileid_18.wav",
         origin + "noises/noise_fileid_6.wav",
         origin + "clean_wav/clean_fileid_18.wav",
         origin + "clean_wav/clean_fileid_6.wav",
        ]
root = "/data/ephraim/datasets/known_noise/undiff_exps/exp_o_net3_6_amplified_{}/"
factors = [3,8,15,20,40,80,120,150,300,500,800]
for f in factors:
    wavs = [root.format(f) + "noisy_wav/book_hfchannel_snr5_fileid_18.wav",
            root.format(f) + "noisy_wav/book_buccaneer1_snr5_fileid_6.wav",
            root.format(f) + "noises/noise_fileid_18.wav",
            root.format(f) + "noises/noise_fileid_6.wav",
            root.format(f) + "clean_wav/clean_fileid_18.wav",
            root.format(f) + "clean_wav/clean_fileid_6.wav",
            ]
    for w in wavs:
        
        wav, sr = torchaudio.load(w)
        print(w)
        if "noises" not in w:
            plot_this(wav, title=w)
        else:
            plot_this(wav[:,:16000], title=w)
        wav = wav*f
        torchaudio.save(w, wav, sr, encoding="PCM_F")
```
