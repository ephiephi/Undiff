```python
import torch

def normalize_estimated_signal(reference: torch.Tensor, estimated: torch.Tensor, eps=1e-8):
    """
    Normalize the estimated signal using SI-SDR-style scaling.
    
    Args:
        reference (Tensor): shape (1, T), reference signal
        estimated (Tensor): shape (1, T), estimated signal
        eps (float): small constant for numerical stability
    
    Returns:
        Tensor: normalized estimated signal (shape (1, T))
    """
    # Remove mean (zero-mean normalization)
    reference = reference - reference.mean(dim=1, keepdim=True)
    estimated = estimated - estimated.mean(dim=1, keepdim=True)
    
    # Compute alpha
    numerator = (estimated * reference).sum(dim=1, keepdim=True)
    denominator = (reference ** 2).sum(dim=1, keepdim=True) + eps
    alpha = numerator / denominator

    # Normalize estimated
    normalized_estimated = estimated / (alpha + eps)

    return normalized_estimated
```
