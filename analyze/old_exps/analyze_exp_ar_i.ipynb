```python
def df_for_snr(snr_path):
    stats_path = os.path.join(snr_path, "stats_dns.json")
    with open(stats_path) as f:
        data = json.load(f)
    df = pd.DataFrame.from_dict(data).T.reset_index()
    for i in range(len(df)):
        df.at[i, "index"] = df["index"][i].split("s")[1]
    df["index"] = df["index"].astype(float)
    df = df.sort_values(by=['index']).reset_index(drop=True)
    df= df.rename(columns={"index": "s"})
    df.to_excel(os.path.join(snr_path,"measurements.xlsx"))

    return df
    
```

```python
from pathlib import Path
from glob import glob
import shutil


def show_results(dir_path, dest):
    dict_ = {"dir":[],"snr": [], "pesq":[], "stoi": [], "s": [] }
    df_results = pd.DataFrame(data=dict_)
    root_ = Path(dir_path)
    sig_dirs = glob(str(root_)+"/*")
    idx = 0
    for j in range(len(sig_dirs)):
        sig = Path(sig_dirs[j])
        if len(sig.stem) >1:
            continue
        src = sig/"clean_wav/"
        src = glob(str(src)+"/*.wav")[0]
        # print(src)
        filename = f"{sig.stem}_clean.wav"
        dst = Path(dest)/filename
        # print(dst)
        shutil.copyfile(src, dst)
        for snr in ["snr5","snr10","snr15"]:
            idx+=1
            
            srcs = glob(str(sig/"noisy_wav/")+"/*.wav")
            print(srcs)
            for src in srcs:
                if snr in src:
                    # print(src)
                    filename = f"{sig.stem}_{snr}_noisy.wav"
                    dst = Path(dest)/filename
                    # print(dst)
                    shutil.copyfile(src, dst)
            
            if sig.stem == "scaling_snrs":
                continue
            df = df_for_snr(Path(sig_dirs[j])/"enhanced_60"/snr)
            max_pesq = float(max(df.pesq))
            c_stoi = float(df[df.pesq == max(df.pesq)]["stoi"].iloc[0])
            # if c_stoi<0.75:
            #     max_pesq_idx = max(df.pesq))
            cur_s = float(df[df.pesq == max(df.pesq)]["s"].iloc[0])
            
            df_results.at[idx, "dir"] = sig.stem

            df_results.at[idx, "snr"] = snr
            df_results.at[idx, "pesq"] = max_pesq
            df_results.at[idx, "stoi"] = c_stoi
            df_results.at[idx, "s"] = cur_s
            # print(Path(sig)/"enhanced60"/str(snr)/("s"+str(cur_s)))
            filename =sig.stem+ "_"+str(snr)+("s"+str(cur_s))+".wav"
            src = Path(sig)/"enhanced_60"/str(snr)/("s"+str(cur_s))
            # print(src)
            # print( glob(str(src)+"/*.wav"))

            src = glob(str(src)+"/*.wav")[0]
            dst = Path(dest)/filename
            # print(src)
            # print(dst)
            shutil.copyfile(src, dst)

            
    return df_results

dirpath = "/data/ephraim/datasets/known_noise/undiff/exp_ar_h_0.99"
df_results = show_results(dirpath,dest="/data/ephraim/datasets/known_noise/undiff/exp_ar_h_0.99/chosen/")
df_results

```
