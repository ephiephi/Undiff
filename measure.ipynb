```python
exp_dir = "/data/ephraim/datasets/known_noise/undiff/exp_ar_i_real/b/"
enhanced_dir =  Path(exp_dir)/"enhanced_60"
clean_dir = Path(exp_dir)/"clean_wav"
noisy_dir =  Path(exp_dir)/"noisy_wav"
snr_dirs = glob(os.path.join(enhanced_dir,"*/"))
```

```python
from tqdm import tqdm
mos_args = argparse.Namespace(
            testset_dir=enhanced_dir, personalized_MOS=False, csv_path=None)
dns_df_all = dnsmos_local.main(mos_args)

data=[]
for snr_dir in tqdm(snr_dirs): 
    # print(snr_dir)
    # snr_dir = "/data/ephraim/datasets/known_noise/enhanced_diffwave_1sec/snr10/"
    stats_path =os.path.join(snr_dir, "new_stats_dns.json")
    # if os.path.exists(stats_path):
    #     os.remove(stats_path)
    s_dirs =glob(snr_dir+"*/")
    # print(s_dirs)
    
    for i,path_ in tqdm(enumerate(s_dirs)):
        wavs_ = glob(path_+"*.wav")
        if len(wavs_)>0:
            ours_files = glob(path_+"*.wav")
            for ours in ours_files:
                # print(glob(path_+"*.wav"))
                enhance, en_sr = torchaudio.load(ours)
                name = Path(ours).name
                clean_path = Path(clean_dir)/name
                speech_wav, _sr = torchaudio.load(clean_path)
                noisy_path =  Path(noisy_dir)/name
                noisy_wav, _sr = torchaudio.load(noisy_path)
                # Compute metrics
                metrics = get_metrics(clean=speech_wav[0].numpy(), mix=noisy_wav[0].numpy(),estimate=enhance[0].numpy(), sample_rate=en_sr)
                metrics["name"] = Path(path_).name
                metrics["filename"] = (ours)
                metrics["snr"] = ours.split("snr")[1].split("/")[0]
                # Print metrics
                # print(metrics)
                data.append(metrics)
    if len(data) >0:
        df = pd.DataFrame.from_records(data,index=range(len(data))) 

print(df[["name","stoi","pesq","si_sdr","sdr", "sir","sar","filename"]])
```

```python
from tqdm import tqdm

noisy_dir = "/data/ephraim/datasets/known_noise/sgmse/exp_i/noisy_synth_i95/"
clean_dir = "/data/ephraim/datasets/known_noise/sgmse/exp_i/clean_synth_i95/"
root = Path("/data/ephraim/datasets/known_noise/sgmse/exp_i/")
models = ["sgmse_WSJ0Chime3", "sgmse_TIMITChime3","storm_TIMITChime3","storm_vbd"]
enhanced_dirs = [root/model_/"enhanced/" for model_ in models]



data=[]
for enhanced_dir in tqdm(enhanced_dirs): 
    mos_args = argparse.Namespace(
            testset_dir=enhanced_dir, personalized_MOS=False, csv_path=None)
    dns_df_all = dnsmos_local.main(mos_args)

    stats_path =os.path.join(enhanced_dir, "measures_storm.json")

    wavs_ = glob(enhanced_dir+"*.wav")
    if len(wavs_)>0:
        ours_files = glob(enhanced_dir+"*.wav")
        for ours in ours_files:
            enhance, en_sr = torchaudio.load(ours)
            name = Path(ours).name
            clean_path = Path(clean_dir)/name
            speech_wav, _sr = torchaudio.load(clean_path)
            noisy_path =  Path(noisy_dir)/name
            noisy_wav, _sr = torchaudio.load(noisy_path)
            # Compute metrics
            metrics = get_metrics(clean=speech_wav[0].numpy(), mix=noisy_wav[0].numpy(),estimate=enhance[0].numpy(), sample_rate=en_sr)
            metrics["name"] = Path(ours).name
            metrics["filename"] = (ours)
            metrics["snr"] = ours.split("snr")[1].split("/")[0]
            if "noise" in Path(ours).name:
                metrics["noise_type"] = ours.split("noise")[1].split("_")[0]
            # Print metrics
            # print(metrics)
            data.append(metrics)
            break
        if len(data) >0:
            df = pd.DataFrame.from_records(data,index=range(len(data))) 
        df = df.merge(dns_df_all, on="filename")
print(df[["name","stoi","pesq","si_sdr","sdr", "sir","sar","filename"]])
```
